```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\user\Documents\Credit Card Application Analysis\clean_credit_application.csv")
```

```python
df.head()
```

```python
print(df.shape)
```

```python
sns.countplot(df['Risk'])
```

```python
fig, axes = plt.subplots(1,3)

plot1= df['Gender'].value_counts().plot.pie(explode=[0.1,0.1], ax=axes[0])
plot1.set_title("Customer Distribution by Gender")

plot2= df['Owning Car'].value_counts().plot.pie(explode=[0.1,0.1], ax=axes[1])
plot2.set_title("Customer Distribution by Car Ownership")

plot3= df['Owning Realty'].value_counts().plot.pie(explode=[0.1,0.1], ax=axes[2])
plot3.set_title("Customer Distribution by Realty Ownership")

fig.set_size_inches(16,6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,20))

cols_to_plot = ['Months', 'Age', 'Years Employed', 'Total Income', 'Number of Children', 'Family Members']
df[cols_to_plot].hist(edgecolor='black', linewidth=1)
fig=plt.gcf()
fig.set_size_inches(15,20)
```

```python
fig, axes = plt.subplots(1,2)

plot1=sns.countplot(y=df['Income Type'], ax=axes[0])
plot1.set_title("Customer Distribution by Income Type")
plot1.set_xlabel("Number of Customers")

plot2=sns.countplot(y=df['Family Status'], ax=axes[1])
plot2.set_title("Customer Distribution by Family Status")
plot2.set_xlabel("Number of Customers")

fig.set_size_inches(15, 6)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,2)

plot1=sns.countplot(y=df['Housing Type'], ax=axes[0])
plot1.set_title("Customer Distribution by Housing Type")
plot1.set_xlabel("Number of Customers")

plot2=sns.countplot(y=df['Education'], ax=axes[1])
plot2.set_title("Customer Distribution by Education Level")
plot2.set_xlabel("Number of Customers")

fig.set_size_inches(15, 6)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,20))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
corr = df.corr().round(4)
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, mask=mask, cmap=cmap)
```
